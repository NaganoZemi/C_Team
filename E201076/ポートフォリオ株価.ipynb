```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
symbols=["3407.JP","7453.JP","5021.JP","6504.JP","8035.JP","8830.JP","3626.JP","4062.JP","3861.JP","4206.JP"]
bigger=[0.072875,0.032264,0.069898,0.001898,0.279055,0.031913,0.282547,0.14916,0.053569,0.02682]
```

```python
start="01/01/17"
end="01/21/23"
```

```python
z=pdr.stooq.StooqDailyReader(symbols=symbols[0],start=start,end=end).read().sort_values(by='Date',ascending=True)
for i in range(0,len(z)):
  for j in range(0,5):
    z.iloc[i,j]=0

for i in range(0,len(symbols)):
  x=pdr.stooq.StooqDailyReader(symbols=symbols[i],start=start,end=end).read().sort_values(by='Date',ascending=True)
  y=x*bigger[i]
  z=z+y
```

```python
z[['Close']]
```

```python
z2=z.dropna()
z2[['Close']]
```

```python
fig=plt.figure(figsize=(12,4))
plt.title("Stock Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True)

plt.plot(z2["Close"])
plt.show
```
